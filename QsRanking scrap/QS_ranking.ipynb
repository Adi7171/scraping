## Convert Scores from 0-100 scale to to 1-5

```python
# def convert(x,a,b,c=1,d=5):
#     """converts values in the range [a,b] to values in the range [c,d]"""
#     ans= round(c + float(x-a)*float(d-c)/(b-a))
#     if ans==0:
#       ans=1

#     return ans
# convert(91,20,100,1,5)
```

```python
new_scores =[]
def convert(x,a,b,c=1,d=5):
    """converts values in the range [a,b] to values in the range [c,d]"""
    ans= round(c + float(x-a)*float(d-c)/(b-a))
    if ans==0:
      ans=1

    return ans

for i in range(len(df['score'])):
  new_scores.append(convert(df['score'][i],20,100,1,5))
```
